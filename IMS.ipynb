**Adding new item**

```python
prod_id=int(input("Enter product id: "))
name=str(input("Enter product name: "))
pr=str(input("Enter product price: "))
types=str(input("Enter product type: "))
m_date=str(input("Enter manufacturing date: "))
e_date=str(input("Enter expiry date: "))
st=int(input("Enter stock: "))

```

```python
record[prod_id]= {"prod_name": name , "price":pr ,"type":types , "manufacturing_date" : m_date , "expiry_date" : e_date , "stock":st}
```

```python
js=json.dumps(record)
fd=open("record.json",'w')
fd.write(js)
fd.close()
```

**Purchasing product**


```python
prod_id=int(input("Enter product id: "))
qty=int(input("Enter quantity: "))

record[prod_id]["stock"]=record[prod_id]["stock"]-qty
```

```python
record
```

**Updating inventory**

```python
js=json.dumps(record)
fd=open("record.json",'w')
fd.write(js)
fd.close()
```

**Sales**

```python
{'prod' : prod_id, 'qn' : qty, 'amount' : record[prod_id]['price'] * qty}
```
